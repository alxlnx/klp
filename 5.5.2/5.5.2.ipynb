```python
import mygraph as mg
import numpy as np
import matplotlib.pyplot as plt

fig = mg.make_figure(dpi_mode='high')
ax = fig.add_subplot()
mg.set_grids(fig)
mg.set_title(ax, r'Калибровочный график зависимости номера канала от энергии $\alpha$-частицы')
mg.set_label(ax, r'Энергия $\alpha$-частицы, МэВ', 'Номер канала')

X = np.array([4.784, 5.490, 6.002, 7.687])
Y = np.array([1968.7, 2257.5, 2561.46, 3137.8])
mg.draw_errorbar(ax, X, Y)
k, s_k = mg.linls(X, Y, through_null=True)
mg.display_linls(ax)
```

```python
def to_energy(N):
    return N / k

dN_Ra = np.array( [30.6, 27.6, 32.3, 31.32] )
print( *[to_energy(dN_Ra[i]) for i in range(len(dN_Ra))] )
Ra = [to_energy(dN_Ra[i]) for i in range(len(dN_Ra))]

N_Am = np.array([1945.66, 2271.83])
print( *[to_energy(N_Am[i]) for i in range(len(N_Am))] )
Am = np.array([to_energy(N_Am[i]) for i in range(len(N_Am))])
dN_Am = [25.44, 24.8]
print( *[to_energy(dN_Am[i]) for i in range(len(dN_Am))] )
dE_Am = np.array([to_energy(dN_Am[i]) for i in range(len(dN_Am))])

print(to_energy(2142.26))
print(to_energy(28.54))

N_U = np.array([1707.153, 1949.01])
print( *[to_energy(N_U[i]) for i in range(len(N_U))] )
U = np.array([to_energy(N_U[i]) for i in range(len(N_U))])
dN_U = np.array([67.757, 48.194])
print( *[to_energy(dN_U[i]) for i in range(len(dN_U))] )
dE_U = np.array([to_energy(dN_U[i]) for i in range(len(dN_U))])
```

```python
print(100*(Ra/X))
print(100*(dE_Am/Am))

print(100*(to_energy(28.54)/to_energy(2142.26)))\

print(100*(dE_U/U))
```

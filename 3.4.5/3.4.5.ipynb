### 3.4.5 Петля гистерезиса (динамический метод)

```python
import pandas as pd
import numpy as np
import mygraph as mg

perm = pd.read_csv('x_y1.txt', sep='\t')
feri = pd.read_csv('x_y2.txt', sep='\t')
iron = pd.read_csv('x_y3.txt', sep='\t')

perm['x,'] *= 15.04/5
perm['y,'] *= 0.88/5

feri['x,'] *= 42/5
feri['y,'] *= 0.07/5

iron['x,'] *= 90.9/5
iron['y,'] *= 0.2/5

fig1 = mg.make_figure(dpi_mode='high')
ax1 = fig1.add_subplot()
mg.set_grids(fig1)
mg.set_title(ax1, 'Начальная кривая намагничивания для пермаллоя')
mg.set_label(ax1, 'H, А/м', 'B, Тл')

#mg.draw_errorbar(ax1, perm['x,'], perm['y,'])
ax1.scatter(perm['x,'], perm['y,'], marker='d', c='k')
```

```python
fig2 = mg.make_figure(dpi_mode='high')
ax2 = fig2.add_subplot()
mg.set_grids(fig2)
mg.set_title(ax2, 'Начальная кривая намагничивания для феррита')
mg.set_label(ax2, 'H, А/м', 'B, Тл')

#mg.draw_errorbar(ax2, feri['x,'], feri['y,'])
ax2.scatter(feri['x,'], feri['y,'], marker='d', c='k')
```

```python
def sigma(dh, sh, db, sb):
    return np.sqrt(
        np.square(sb/dh) +\
        np.square(db*sh/dh/dh)
    )
u0 = 4*np.pi/1e7
sigma(15.3, 0.1, 0.09, 0.003) / u0
sigma(0.63, 0.07, 0.268, 0.0025) / u0
sigma(8.3, 0.1, 0.014, 0.0002) / u0
sigma(7.93, 0.17, 0.03, 0.0004) / u0
sigma(18.2, 0.2, 0.041, 0.0005) / u0
sigma(18.4, 0.32, 0.1, 0.0009) / u0
```

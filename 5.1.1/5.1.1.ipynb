```python
import numpy as np
import pandas as pd
import mygraph as mg

calibr = pd.read_csv('calibr.csv', sep='\t')
V0_s = []

fig = mg.make_figure(dpi_mode='high')
ax = fig.add_subplot()
mg.set_grids(fig)
mg.set_title(ax, 'Калибровочная зависимость длины волны падающего на фотокатод света от показаний барабана')
mg.set_label(ax, 'Градус на барабане', 'Длина волны, Å')
#k, s_k, b, s_b = mg.linls(np.array(df['n']), np.array(Ts))
mg.draw_errorbar(ax, calibr['deg'], calibr['WL'], x_err=5, y_err=0)
#mg.display_linls(ax)
```

```python
WL1 = pd.read_csv('WL1.csv', sep='\t')
WL1 = WL1.sort_values(by='V')

fig1 = mg.make_figure(dpi_mode='high')
ax1 = fig1.add_subplot()
mg.set_grids(fig1)
mg.set_title(ax1, r'Зависимость $\sqrt{V_I}$ от V для 22 линии (5852Å)')
mg.set_label(ax1, r'Напряжение на фотоэлементе, В', r'Корень из напряжения, пропорционального фототоку, $ \sqrt{В} $')
#X = WL1['V'][WL1['V_I'] > 0]
#Y = np.sqrt( WL1['V_I'][WL1['V_I'] > 0] )
X = WL1['V']
Y = np.append( -np.sqrt( -WL1['V_I'][WL1['V_I'] < 0] ), np.sqrt( WL1['V_I'][WL1['V_I'] > 0] ) )

mg.draw_errorbar(ax1, X, Y, x_err=0.01, y_err=0.01)

FRT_OFFSET = 7
BCK_OFFSET = 15
#X = X.to_numpy()[:-OFFSET]
#Y = Y.to_numpy()[:-OFFSET]
X = X.to_numpy()[FRT_OFFSET:-BCK_OFFSET]
Y = Y[FRT_OFFSET:-BCK_OFFSET]
k, s_k, b, s_b = mg.linls(X, Y)
mg.display_linls(ax1)
V0_s.append(b)
```

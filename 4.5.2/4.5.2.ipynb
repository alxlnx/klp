```python
import mygraph as mg
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('1.tsv', sep='\t')
df
```

```python
d = df['h1']/df['h2']
nu_1 = 2*np.sqrt(d)/(1+d)

nu = df['h4'] - df['h3'] / (df['h4'] + df['h3'])

nu3 = nu/nu_1

fig = mg.make_figure(dpi_mode='high')
ax = fig.add_subplot()
mg.set_grids(fig)
mg.set_title(ax, r'Зависимость $\nu_3$ от $cos\beta$')
mg.set_label(ax, r'$cos\beta$ ', r'$\nu_3$')

X = np.cos(df['b, deg'] / 360 * 2 * np.pi)
Y = nu3
mg.draw_errorbar(ax, X, Y, x_err=0.02, y_err=.6)
```

```python
fig = mg.make_figure(dpi_mode='high')
ax = fig.add_subplot()
mg.set_grids(fig)
mg.set_title(ax, r'Зависимость $\nu_3$ от $cos^2\beta$')
mg.set_label(ax, r'$cos^2\beta$ ', r'$\nu_3$')

X = np.square( np.cos(df['b, deg'] / 360 * 2 * np.pi) )
Y = nu3
mg.draw_errorbar(ax, X, Y, x_err=0.02, y_err =.6)
```

```python
df = pd.read_csv('2.tsv', sep='\t')
df

```

```python
fig = mg.make_figure(dpi_mode='high')
ax = fig.add_subplot()
mg.set_grids(fig)
mg.set_title(ax, r'Зависимость $\nu_2$ от x')
mg.set_label(ax, r'$x, см$ ', r'$\nu_2$')

d = df['h1'] / df['h2']
nu_1 = 2 * np.sqrt(d) / (1 + d)
nu = df['h4'] - df['h3'] / (df['h4'] + df['h3'])
nu_2 = nu / nu_1

X = df['x, cm']
Y = nu_2
mg.draw_errorbar(ax, X, Y, x_err=1, y_err=.6)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6,6))

ax.grid(which='major', linestyle='--')
ax.grid(which='minor', linestyle=':')
ax.minorticks_on()

g = 9.815
L = 223 / 1e2
M = 2905 / 1e3
def u(mass, delta_x):
    mass /= 1e3
    delta_x /= 1e3
    return (M / mass) * delta_x * np.sqrt(g / L)

masses =   [0.500, 0.513, 0.510, 0.512]
delta_xs = [11, 11.5, 10.75, 11.5]
us = []
for mass, delta_x in zip(masses, delta_xs):
    current_u = u(mass, delta_x)
    us.append(current_u)
print(us)
```
